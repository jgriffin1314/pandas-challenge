```python
# Dependencies and Setup
import pandas as pd
import numpy as np

from pandas import DataFrame
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

df = school_data_complete
merge_df = df
df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_df = df.groupby('school_name')['school_name'].nunique()
school_count = school_df.count()
print("Total school count is " + str(school_count))

student_df = df.groupby('Student ID')['Student ID'].nunique()
student_count = student_df.count()
print("Total student count is " + str(student_count))

budget_df = df.groupby('school_name')['budget'].mean()
total_budget = budget_df.sum()
print("Total budget is " + str(total_budget))

average_math_score = round(df['math_score'].mean(),2)
print("Average math score is " + str(average_math_score))

average_reading_score = round(df['reading_score'].mean(),2)
print("Average reading score is " + str(average_reading_score))

passing_math_df = df[df['math_score'] >= 70]
passing_math_count = passing_math_df['math_score'].count()
passing_math_percentage = round(passing_math_count / student_count * 100,2)
print("Passing math percentage is " + str(passing_math_percentage))

passing_reading_df = df[df['reading_score'] >= 70]
passing_reading_count = passing_reading_df['reading_score'].count()
passing_reading_percentage = round(passing_reading_count / student_count * 100,2)
print("Passing reading percentage is " + str(passing_reading_percentage))


df.loc[(df['math_score'] >= 70) & (df['reading_score'] >= 70), 'passing_both'] = True
passing_both_df = df[df['passing_both'] == True]
passing_both_count = passing_both_df['passing_both'].count()
passing_both_percentage = round(passing_both_count / student_count * 100,2)
print("Passing both percentage is " + str(passing_both_percentage))

data = { "School Count": [school_count],
       "Student Count": [student_count],
       "Total Budget": [total_budget],
       "Average Math Score": [average_math_score],
       "Average Reading Score": [average_reading_score],
       "% Passing Math": [passing_math_percentage],
       "% Passing Reading": [passing_reading_percentage],
       "% Passing Both": [passing_both_percentage]}


clean_df = pd.DataFrame(data)
clean_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Students by school
total_students_by_school_df = df.groupby('school_name')['Student ID'].count()
total_students_by_school = total_students_by_school_df

#Budget by school
total_budget_by_school_df = df.groupby('school_name')['budget'].mean()
total_budget_by_school = total_budget_by_school_df

#Average math score by school
math_score_by_school_df = df.groupby('school_name')['math_score'].mean()
math_score_by_school = math_score_by_school_df
math_score_by_school = round(math_score_by_school,2)

#Average reading score by school
reading_score_by_school_df = df.groupby('school_name')['reading_score'].mean()
reading_score_by_school = reading_score_by_school_df
reading_score_by_school = round(reading_score_by_school,2)

#% passing math by school
passing_math_df = df[(df["math_score"] >= 70)]
passing_math_by_school_df = passing_math_df.groupby('school_name')['Student ID'].count()
passing_math_by_school_count = passing_math_by_school_df

#% passing reading by school
passing_reading_df = df[(df["reading_score"] >= 70)]
passing_reading_by_school_df = passing_reading_df.groupby('school_name')['Student ID'].count()
passing_reading_by_school_count = passing_reading_by_school_df

#% passing both by school
passing_both_df = passing_math_df[(passing_math_df["reading_score"] >= 70)]
passing_both_by_school_df = passing_both_df.groupby('school_name')['Student ID'].count()
passing_both_by_school_count = passing_both_by_school_df

#dataframe summary
school_summary = {"Total Students": total_students_by_school,
                  "Total Budget": total_budget_by_school,
                  "Per Student Budget" : total_budget_by_school / total_students_by_school,
                  "Average Math Score": math_score_by_school,
                  "Average Reading Score" : reading_score_by_school,
                  "% Passing Math": round(passing_math_by_school_count / total_students_by_school*100,2),
                  "% Passing Reading": round(passing_reading_by_school_count / total_students_by_school*100,2),
                  "% Passing Both": round(passing_both_by_school_count / total_students_by_school*100,2),

                 }
school_summary_df = pd.DataFrame(school_summary)
school_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 585, 630, 645, 680]
group_names = ['<$585','$585-630', '$630-645', '$645-680']

school_summary_df["Budget Group"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_names)
school_summary_df

Average_math_score_by_budget = round(school_summary_df.groupby('Budget Group')['Average Math Score'].mean(),2)

Average_reading_score_by_budget = round(school_summary_df.groupby('Budget Group')['Average Reading Score'].mean(),2)

Percent_passing_math_by_budget = round(school_summary_df.groupby('Budget Group')['% Passing Math'].mean(),2)

Percent_passing_reading_by_budget = round(school_summary_df.groupby('Budget Group')['% Passing Reading'].mean(),2)

Percent_passing_both_by_budget = round(school_summary_df.groupby('Budget Group')['% Passing Both'].mean(),2)

by_budget_dict = {'Average Math Score': Average_math_score_by_budget,
     'Average Reading Score': Average_reading_score_by_budget,
     '% Passing Math': Percent_passing_math_by_budget,
     '% Passing Reading': Percent_passing_reading_by_budget,
     '% Passing Both': Percent_passing_both_by_budget,
    }
averages_by_size_df = pd.DataFrame(data=by_budget_dict)
averages_by_size_df
```
